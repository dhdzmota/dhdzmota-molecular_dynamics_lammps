```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib notebook

# Parameters for beautifulize the plots
plt.rcParams.update({'mathtext.default':  'regular' })
sns.set(context='paper',style="whitegrid",font='Times New Roman', font_scale=1.2)
```

```python
with open ('./FeOH3-rdf0300.dat', 'r') as fout:
    value=fout.read()
text=value.split('\n20000 100\n')[1]
```

```python
fdr = pd.DataFrame([i.split(' ')for i in text.split('\n')][:-1],dtype=float,
                   columns = ['Num','D',
                              'gr11','int_gr11',
                              'gr22','int_gr22',
                              'gr33', 'int_gr33',
                              'gr12','int_gr12',
                              'gr13','int_gr13',
                              'gr23', 'int_gr23',])

fig=plt.figure()
plt.plot(fdr.D, fdr.gr33, label='Fe-Fe')
plt.plot(fdr.D, fdr.gr11, label='H-H')
plt.plot(fdr.D, fdr.gr22, label='O-O')
plt.legend()

plt.show()
```

```python
fig=plt.figure()
plt.plot(fdr.D, fdr.gr12, label='H-O')
plt.plot(fdr.D, fdr.gr13, label='H-Fe')
plt.plot(fdr.D, fdr.gr23, label='O-Fe')
plt.legend()
```
